```python
# Task 2 — Cart Operations 
cart = []


def print_cart(cart):
    print("Current Cart:")
    if not cart:
        print("Your cart is empty.")
    else:
        for item in cart:
            print(f"- {item['name']} (x{item['quantity']})")    
    print() # Adds a gap after cart display



# to update the cart if item is in menu and available, if item already in cart, increase quantity, does not exist in menu or unavailable, print appropriate message..
def add_item(cart, item_name, quantity):
    if item_name not in menu:
        print(f"Sorry, '{item_name}' is not on the menu.")
        return
    if not menu[item_name]["available"]:
        print(f"Sorry, '{item_name}' is currently unavailable.")
        return
    
    for item in cart:
        if item["name"] == item_name:
            item["quantity"] += quantity
            print(f"Updated '{item_name}' quantity to {item['quantity']} in the cart.")
            return
    
    cart.append({"name": item_name, "quantity": quantity})
    print(f"Added '{item_name}' (x{quantity}) to the cart.")

add_item(cart, "Paneer Tikka", 2)
add_item(cart, "Chicken Wings", 1)
add_item(cart, "Veg Soup", 1)
add_item(cart, "Butter Chicken", 3)
add_item(cart, "Dal Tadka", 2)
add_item(cart, "Veg Biryani", 1)
add_item(cart, "Garlic Naan", 4)
add_item(cart, "Gulab Jamun", 2)
add_item(cart, "Curd Rice", 1)

print(f"-----------------")

#Q2. Write logic to remove an item from the cart by name.
def remove_item(cart, item_name):
    for item in cart:
        if item["name"] == item_name:
            cart.remove(item)
            print(f"Removed '{item_name}' from the cart.")
            return
    print(f"'{item_name}' not found in the cart.")
remove_item(cart, "Veg Soup")
remove_item(cart, "Ice Cream")


print(f"-----------------")

#Q3. Write logic to update the quantity of an item already in the cart.

def update_quantity(cart, item_name, new_quantity):
    for item in cart:
        if item["name"] == item_name:
            item["quantity"] = new_quantity
            print(f"Updated '{item_name}' quantity to {new_quantity} in the cart.")
            return
    print(f"'{item_name}' not found in the cart.")
update_quantity(cart, "Butter Chicken", 5)
update_quantity(cart, "Gulab Jamun", 3)
update_quantity(cart, "Rasgulla", 2)
update_quantity(cart, "Ice Cream", 10)

#Q4.Using your logic above, simulate the following sequence. Print the cart state after each step:
#Add "Paneer Tikka" × 2

#Step 1: Add "Paneer Tikka" × 2 
print("Step 1: Add 'Paneer Tikka' × 2")
add_item(cart, "Paneer Tikka", 2)
print_cart(cart)
#Add "Gulab Jamun" × 1 (should show unavailable)
print("Step 2: Add 'Gulab Jamun' × 1")
add_item(cart, "Gulab Jamun", 1)
print_cart(cart)

#Add "Paneer Tikka" × 1
print("Step 3: Add 'Paneer Tikka' × 1")
add_item(cart, "Paneer Tikka", 1)
print_cart(cart)

#Add "Mystery Burger" × 1 (not on menu)
print("Step 4: Add 'Mystery Burger' × 1")
add_item(cart, "Mystery Burger", 1)
print_cart(cart)

#Add "Chicken Wings" × 1 (unavailable)
print("Step 5: Add 'Chicken Wings' × 1")
add_item(cart, "Chicken Wings", 1)
print_cart(cart)
#Remove "Gulab Jamun" (not in cart)
print("Step 6: Remove 'Gulab Jamun' (not in cart)")
remove_item(cart, "Gulab Jamun")
print_cart(cart)

#Q4. Print final order Summary:
def print_order_summary(cart):
    width = 20
    print("="*50)
    print(f"order summary".center(50))
    print("="*50)

    subtotal = 0.0
    for item in cart:
        name = item["name"]
        quantity = item["quantity"]
        price = menu[name]["price"]
        total_price = price * quantity
        subtotal += total_price
        print(f"{name:<{width}} x{quantity:<3} ₹{total_price:.2f}")

    gst = 0.18 * subtotal
    total = subtotal + gst

    print("-"*50)
    print(f"{'Subtotal':<{width}} ₹{subtotal:.2f}")
    print(f"{'GST (18%)':<{width}} ₹{gst:.2f}")
    print("-"*50)
    print(f"{'Total':<{width}} ₹{total:.2f}")
    print("="*50)
print_order_summary(cart)
```

```python
#Task 3 — Inventory Management

#Deep copy inventory into a variable called inventory_backup before making any changes.To demonstrate this works, manually change one stock value in inventory and print both inventory and inventory_backup to show that the backup is unaffected. Then restore inventory to its original state before continuing.(Hint: import copy and use copy.deepcopy().)
import copy
inventory_backup = copy.deepcopy(inventory)
# Manually change one stock value in inventory
inventory["Paneer Tikka"]["stock"] = 5

print("Inventory after change:")
print(inventory)
print("\nInventory Backup (should be unaffected):")
print(inventory_backup)
# Restore inventory to its original state
inventory = copy.deepcopy(inventory_backup)
print("\nInventory after restoration:")
print(inventory)

print("-----------------")

#Q2. Simulate order fulfilment: deduct the quantities from the final cart in Task 2 from the corresponding items in inventory.If stock for an item is insufficient, print a warning and deduct only what is available (do not go below 0).

def fulfill_order(cart, inventory):
    for item in cart:
        name = item["name"]
        quantity = item["quantity"]
        if name in inventory:
            available_stock = inventory[name]["stock"]
            if available_stock >= quantity:
                inventory[name]["stock"] -= quantity
                print(f"Fulfilled '{name}' x{quantity}. Remaining stock: {inventory[name]['stock']}")
            else:
                inventory[name]["stock"] = 0
                print(f"Warning: Only {available_stock} '{name}' available. Fulfilled {available_stock} and set stock to 0.")
        else:
            print(f"Warning: '{name}' not found in inventory. Cannot fulfill this item.")
fulfill_order(cart, inventory)


#Q3 After deduction, loop through inventory and print a Reorder Alert for every item whose stock is at or below its reorder_level:⚠ Reorder Alert: Paneer Tikka — Only 7 unit(s) left (reorder level: 3)
print("\nReorder Alerts:")
for item_name, details in inventory.items():    
    if details["stock"] <= details["reorder_level"]:
        print(f"⚠ Reorder Alert: {item_name} — Only {details['stock']} unit(s) left (reorder level: {details['reorder_level']})")



#Q4. Print both inventory and inventory_backup at the end to confirm they differ — proving the deep copy protected the original.
print("\nFinal Inventory:")
print(inventory)
print("\nInventory Backup (should be unchanged):")
print(inventory_backup)

```

```python
#Task 4 — Daily Sales Log Analysis

#Q1. Print total revenue per day.
print("Total Revenue Per Day:")
for date, orders in sales_log.items():
    total_revenue = sum(order["total"] for order in orders)
    print(f"{date}: ₹{total_revenue:.2f}")

#Q2. Print the best-selling day (date with the highest total revenue).
best_selling_day = max(sales_log.items(), key=lambda item: sum(order["total"] for order in item[1]))
print(f"\nBest-Selling Day: {best_selling_day[0]} with total revenue of ₹{sum(order['total'] for order in best_selling_day[1]):.2f}")

#Q3. Find the most ordered item — the item that appears in the greatest number of individual orders across all days.
item_order_count = {}
for orders in sales_log.values():
    for order in orders:
        for item in order["items"]:
            item_order_count[item] = item_order_count.get(item, 0) + 1

most_ordered_item = max(item_order_count, key=item_order_count.get)
print(f"\nMost Ordered Item: {most_ordered_item} with {item_order_count[most_ordered_item]} orders")

#Q4.Add the following new day to sales_log and then reprint the revenue-per-day table and best-selling day to show the stats update correctly:
sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"],                     "total": 260.0},
]
print("\nUpdated Total Revenue Per Day:")
for date, orders in sales_log.items():
    total_revenue = sum(order["total"] for order in orders)
    print(f"{date}: ₹{total_revenue:.2f}")
best_selling_day = max(sales_log.items(), key=lambda item: sum(order["total"] for order in item[1]))
print(f"\nUpdated Best-Selling Day: {best_selling_day[0]} with total revenue of ₹{sum(order['total'] for order in best_selling_day[1]):.2f}")   

#Q5.Using enumerate, print a numbered list of all orders across all dates (including the new day):
print("\nAll Orders:")
order_number = 1
for date, orders in sales_log.items():
    for order in orders:
        print(f"{order_number}. Order ID: {order['order_id']}, Total: ₹{order['total']:.2f}")
        order_number += 1
        
```
